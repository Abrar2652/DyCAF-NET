```python
import pandas as pd
import numpy as np

# Load the results CSV file
results_file = '/kaggle/input/spikenet/Y8_SpNet(TD).csv'  # Replace with the actual path
results_df = pd.read_csv(results_file)

# Extract the relevant columns for precision, recall, mAP@50, mAP@50-95 from the CSV
precision_values = results_df['metrics/precision(B)']
recall_values = results_df['metrics/recall(B)']
map_50_values = results_df['metrics/mAP50(B)']
map_50_95_values = results_df['metrics/mAP50-95(B)']

# Calculate the mean and standard deviation for each metric
precision_mean = np.mean(precision_values)
precision_std = np.std(precision_values)

recall_mean = np.mean(recall_values)
recall_std = np.std(recall_values)

map_50_mean = np.mean(map_50_values)
map_50_std = np.std(map_50_values)

map_50_95_mean = np.mean(map_50_95_values)
map_50_95_std = np.std(map_50_95_values)

# Print the results in the desired format
print(f"Precision - Mean: {precision_mean:.4f}±{precision_std:.4f}")
print(f"Recall - Mean: {recall_mean:.4f}±{recall_std:.4f}")
print(f"mAP@50 - Mean: {map_50_mean:.4f}±{map_50_std:.4f}")
print(f"mAP@50-95 - Mean: {map_50_95_mean:.4f}±{map_50_95_std:.4f}")
```
